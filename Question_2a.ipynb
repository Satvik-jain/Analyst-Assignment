```python
import pandas as pd
import numpy as np
from math import cos,sin,pi
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv('dataset_invst.csv')
```

```python
d
```

```python
x = d['close']
y = d['ema_fast']
z = d['ema_slow']
t = d['datetime']
```

```python
plt.plot(t,y,color="r",label="Ema_fast")
plt.plot(t,x,color="g",ls="dotted", label= "Close")
plt.plot(t,z,color="b",label="Ema_slow")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Closing price of the stock along with the EMA_fast and EMA_slow")
ax = plt.gca()
ax.set_xticklabels([])
plt.show()
```

```python
daily_returns = ((d['close']-d['open'])/d['open'])*100
plt.hist(daily_returns, 200,
         histtype='bar',
         facecolor='b',
         alpha=0.5) # alpha is opacity
plt.xlabel("% returns")
plt.ylabel("Frequency")
plt.title("Segregating daily returns")
plt.show()
```

```python
daily_returns = d['close']-d['open']
red_candles = 0
green_candles = 0
for candle in daily_returns:
  if candle>=0:
    green_candles+=1
  else:
    red_candles+=1
print(f'The number of green candles are {green_candles}')
print(f'The number of red candles are {red_candles}')
```

```python
def count_three_white_soldiers(x):
  count = 0
  for index in range(2,len(x)):
    if x[index]>0 and x[index-1]>0 and x[index-2]>0:
      count += 1
  return count

def count_three_black_crows(x):
  count = 0
  for index in range(1,len(x)):
    if x[index]<0 and x[index-1]<0 and x[index-2]<0:
      count += 1
  return count
print(f'The number of Three white soldiers are {count_three_white_soldiers(daily_returns)}')
print(f"The number of Three black crows are {count_three_black_crows(daily_returns)}")
```
